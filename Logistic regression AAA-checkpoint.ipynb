```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_excel('air cargo supplychain.xlsx')
```

```python
if "Shipment ID" in df.columns:
    df = df.drop(columns=["Shipment ID"])
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Handle missing values (if any) using mean for numerical and mode for categorical
num_imputer = SimpleImputer(strategy="mean")
cat_imputer = SimpleImputer(strategy="most_frequent")
```

```python
df[numerical_cols] = num_imputer.fit_transform(df[numerical_cols])
df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])
```

```python
# Split dataset into features and target variable
X = df.drop(columns=["Delay Status"])  # Features
y = df["Delay Status"]  # Target
```

```python
# Standardize numerical features
scaler = StandardScaler()
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])
```

```python
# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2000)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Train Logistic Regression Model
log_reg = LogisticRegression(max_iter=500)
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)
```

```python
# Train Decision Tree Classifier Model
dt_clf = DecisionTreeClassifier(random_state=45)
dt_clf.fit(X_train, y_train)
y_pred_dt = dt_clf.predict(X_test)
```

```python
# Evaluate models
log_reg_acc = accuracy_score(y_test, y_pred_log)
dt_clf_acc = accuracy_score(y_test, y_pred_dt)
```
